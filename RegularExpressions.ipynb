# 5)DIY Pattern Matching - Continuous Numbers

```python
#Now let us write our own logic to find out the patterns without using regular expressions
```

```python
txt = 'The number is 8372493 not 423423.'

c = 0

for i in range(len(txt)):
    if (txt[i].isdigit() == True):        # If Digit
        if (c == 0):                      # Count Number of Digits
            st = i                        # Starting the Number
        c += 1                            # Incrementing the Counter for that number
    else:                                 # If not Digit
        if (c != 0):                      # If coming from a digit
            end =  i                      # Ending the Number
            print(txt[st:end])            # Printing the Number
        c = 0

if (c != 0):                              # If Number is at the end of the string
    print(txt[-c:])
```

# 7)DIY Pattern Matching - Words starting from specific characters

```python
text = 'AI is having the capability to revolutionize all the industries. The Industry is ready for a big change'
word = 'in'
```

- 1.words Starting with Specific Characters

```python
for wrd in text.split(' '):
  if(wrd[:2].lower() == word.lower()):
    print(wrd)
```

- 2.Words Ending with Specific Characters

```python
text = 'AI is having the capability to revolutionize all the industries. The Industry is ready for a big change'
word = 'e'
for wrd in text.split(' '):
  if(wrd[-1:].lower() == word.lower()):
    print(wrd)
```

# 8)DIY Pattern Matching - Email Attraction

```python
mails = ['ashish@gmail.com','ashish@geeksforgeeks.org',
         'ashish@orkut.com','ashish@yahoo.com',
         'ashish.jangra@geeksforgeeks.org','ashish@gfg.co.in',
         'ashish@aap.gov.in','aj@gmail.co.org.in']
```

```python
user_id = []
host_name = []
domain_type = []

for mail in mails:
    user_id.append(mail.split('@')[0])
    host_name.append(mail.split('@')[1].split('.')[0])
    domain_type.append('.'.join(mail.split('@')[1].split('.')[1:]))
```
