# 1)Making the Pairs | Uni-Gram

```python
import re
from tqdm import tqdm
import numpy as np
import pandas as pd
```

```python
with open('big.txt' , 'r') as fd:
    lines = fd.readlines()

    words = []
    for word in lines:
        words += re.findall('\w+' , word.lower())
```

```python
words
```

```python
data = []
for i in range(len(words)-1):
    data.append(words[i:i+2])
```

```python
data
```

```python
#we need to join the words 
def get_pairs(words):
    data = []
    for i in range(len(words)-1):
        data.append(' '.join(words[i:i+2]))
    return data
```

```python
data = get_pairs(words)
```

```python
data
```

# 3)Predicting the Words

```python
df = pd.DataFrame(probability_dist , columns = ['pair' , 'freq' , 'out'])
df.head()
```

```python
df
```

```python
#remove all the pairs which are less than 5
df = df[df['freq'] >=5]
```

```python
df.head()
```

```python
df_pred = []

word = 'he' 
for i in df.values:
    if i[0].split(' ')[0] == word:
        df_pred.append([i[0] , i[1], i[2]])

df_pred = pd.DataFrame(df_pred , columns = ['in' , 'freq' , 'out'])
print(list(df_pred.sort_values(by = 'freq' , ascending = False).head()['out'].values))
```

```python
## We will pass this in a whole function 
df = pd.DataFrame(probability_dist , columns = ['pair' , 'freq' , 'out'])
df = df[df['freq'] >=5]
df.head()

word = 'the' 
def predict(word):

    df_pred = []
    
    for i in df.values:
        if i[0].split(' ')[0] == word:
            df_pred.append([i[0] , i[1], i[2]])
    df_pred = pd.DataFrame(df_pred , columns = ['in' , 'freq' , 'out'])
    return list(df_pred.sort_values(by = 'freq' , ascending = False).head()['out'].values)
    
```
