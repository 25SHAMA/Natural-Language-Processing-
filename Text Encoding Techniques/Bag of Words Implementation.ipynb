# 1. Bag of Words

```python
#In one hot encoding we do not have frequency but in bag of words we are going to implement frequency
feature_vectors = []
text = ["This is a very very cute cat", "The dog is very cute"]

tokens = [token.split() for token in text]
print(tokens)
```

```python
vocab = list(set([word for text in tokens for word in text]))
```

```python
vocab
```

```python
for text in tokens:
    feature_vector = [0]*len(vocab)
    for word in text:
        feature_vector[vocab.index(word)] = text.count(word)
    feature_vectors.append(feature_vector)   
```

```python
print(vocab)
print(feature_vectors)
```

```python
import pandas as pd
pd.DataFrame(feature_vectors, columns = vocab)
```

# 2. Binary Bag of Words

```python
feature_vectors = []
text = ["This is a very very cute cat", "The dog is very cute"]

tokens = [token.split() for token in text]
print(tokens)
vocab = list(set([word for text in tokens for word in text]))
for text in tokens:
    feature_vector = [0]*len(vocab)
    for word in text:
        if word in vocab:
            feature_vector[vocab.index(word)] = 1
    feature_vectors.append(feature_vector)   
print(vocab)
print(feature_vectors)#binary bag of words and one hot encoding is same but bag of words is different in which we have frequency
```
